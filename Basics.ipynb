```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

fig = plt.figure(figsize=(18, 36))

###############pots first column based on stem_cats list
for sp in range(0,6):
    ax = fig.add_subplot(6,3,sp*3+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], 
            c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], 
            c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2004, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
        
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

fig = plt.figure(figsize=(18, 36))
#By the way, this is pretty terrible code! It has a lot of repetition
#and quite frankly should use more lists and the power of the
#loop to avoid repeating so many commands. Yet this is a learning
#experience, so I will leave it as it is for now!

###########plots the stem_cats list
for sp in range(0,6):
    #generates a subplot in a 6x3 area. Labels each chart
    #from left to right 1, 2, 3, etc. Only prints "sp*3+1"th chart
    ax = fig.add_subplot(6,3,sp*3+1)
    #choose which data to plot
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], 
            c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], 
            c=cb_orange, label='Men', linewidth=3)
    #removes right, left, top, and bottom spines
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    #sets the max x and y limits
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    #sets the title of each chart to the name from the list
    ax.set_title(stem_cats[sp])
    #turns off tick parameters on bottom, top, etc
    ax.tick_params(labelbottom='off', bottom="off", top="off", left="off", right="off")
    #sets min and max y values
    ax.set_yticks([0,100])
    #creates a line across the middle (50) with a gray color
    #and 0.3 transparency
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    
    #selects only the first and last plot and creates text to
    #label the lines Men or Women. Finally, removes the 
    #labels for the x and y charts
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
        ax.tick_params(labelbottom='on')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2005, 35, 'Women')
        ax.tick_params(labelbottom='on')
        
#########plots the 3rd column
    ax = fig.add_subplot(6,3,sp*3+3)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[sp]], 
            c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[sp]], 
            c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(other_cats[sp])
    ax.tick_params(labelbottom='off', bottom="off", top="off", left="off", right="off")
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    
    if sp == 0:
        ax.text(2005, 75, 'Women')
        ax.text(2005, 20, 'Men')
        ax.tick_params(labelbottom='on') 
    elif sp == 5:
        ax.text(2005, 65, 'Men')
        ax.text(2005, 33, 'Women')
        ax.tick_params(labelbottom='on')
        
#########plots the 2nd column with lib_arts_cats list
for sp in range(0,5):
    ax = fig.add_subplot(6,3,sp*3+2)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[sp]],c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[sp]],c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(lib_arts_cats[sp])
    ax.tick_params(labelbottom='off', bottom="off", top="off", left="off", right="off")
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    
    if sp == 0:
        ax.text(2005, 75, 'Women')
        ax.text(2007, 22, 'Men')
        ax.tick_params(labelbottom='on')
    elif sp == 4:
        ax.text(1975, 70, 'Men')
        ax.text(1975, 27, 'Women')
        ax.tick_params(labelbottom='on')

#plots, saves, and exports the graphs        
plt.plot(women_degrees['Year'], women_degrees['Biology'])
plt.savefig('gender_degrees.png')        
```
